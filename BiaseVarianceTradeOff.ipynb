In this Notebook I compare the Mean-Squared Error (MSE) for two estimators of the parameter $\theta$, where observations $X_1,X_2,\cdots,X_n$~Uniform$(\theta)$. The first estimator is:
$$
\hat\theta_n = \max (X_1,X_2,\cdots,X_n)
$$
and the other estimator is
$$
\hat \beta_n = 2 \bar{X_n}
$$
where $\bar{X_n}$ is the average of the observations. One can show that for the first estimator:
$$
\text{bias}(\hat\theta_n) = -\frac{\theta}{n+1} \ \ \  \text{and}\ \ \ \text{var}(\hat\theta_n) = \theta^2\big[\frac{n}{n+2} - \frac{n^2}{(n+1)^2}\big] 
$$
leading to the MES of:
$$
\text{MSE}(\hat\theta_n) = \big[\frac{\theta}{n+1}\big]^2 + \theta^2\big[\frac{n}{n+2} - \frac{n^2}{(n+1)^2}\big]
$$
Similarly, the bias and variance for the second estimator are
$$
\text{bias}(\hat\beta_n) = 0 \ \ \  \text{and}\ \ \ \text{var}(\hat\beta_n) = \frac{\theta^2}{3n} 
$$
respectively leading to the overall MSE of:
$$
\text{MSE}(\hat\beta_n) = \frac{\theta^2}{3n} 
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
n = np.linspace(10,500,20)
theta = 1000
```

```python
# First Estimator
bias_th = -theta/(n+1)
var_th = (theta**2)*(n/(n+2) - n**2/(n+1)**2)
MSE_th = bias_th**2 + var_th
TH = pd.DataFrame({'Estimator': 'Theta', 'Bias': [bias_th], 'Variance': [var_th], 'MSE': [MSE_th]})
```

```python
TH
```

```python
# First Estimator
bias_b = 0/(n+1)
var_b = (theta**2)/(3*n)
MSE_b = bias_b**2 + var_b
B = pd.DataFrame({'Estimator': 'Beta', 'Bias': [bias_b], 'Variance': [var_b], 'MSE': [MSE_b]})
```

```python
B
```

```python
# Plot for Theta
plt.figure(figsize = (10,10))
plt.title('Theta')
plt.grid
plt.xlabel('n')
Labels = ['Bias','Variance','MSE']
for i in Labels:
    plt.plot(n,TH.loc[0,i])
plt.legend(Labels, ncol = 3, loc = 'upper right')
plt.show()
```

```python
# Plot for Beta
plt.figure(figsize = (10,10))
plt.grid
plt.title('Beta')
plt.xlabel('n')
Labels = ['Bias','Variance','MSE']
for i in Labels:
    plt.plot(n,B.loc[0,i])
plt.legend(Labels, ncol = 3, loc = 'upper right')
plt.show()
```

```python
# Now compare the MSE for the two side by side
plt.figure(figsize = (10,10))
plt.grid
plt.title('Theta vs Beta')
plt.xlabel('n')
Labels = ['Theta','Beta']
for i in Labels:
    if "h" in i:
        plt.plot(n,TH.loc[0,'MSE'])
    else:
        plt.plot(n,B.loc[0,'MSE'])   
plt.legend(Labels, ncol = 2, loc = 'upper right')
plt.show()
```
